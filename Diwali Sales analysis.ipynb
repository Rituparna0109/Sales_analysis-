ANALYSIS

```python
ax=sns.countplot(x='Gender', data= df)
```

```python
ax=sns.countplot(x='Gender', data= df)
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
sale_gen= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by ='Amount' , ascending=False)
sns.barplot(x='Gender',y= 'Amount', data= sale_gen )
```

NOTE-As we can see most buyers are Females and purchasing power is also greater then men

```python
#age bases
#hue divide in gender
ax=sns.countplot(x='Age Group', data= df, hue='Gender')
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
sale_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by ='Amount' , ascending=False)
sns.barplot(x='Age Group',y= 'Amount', data= sale_age )
```

NOTE-Most of buyers are from age 26-35 and they are female

```python
#STATE
sale_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by ='Orders' , ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(x='State',y= 'Orders', data= sale_state )
```

```python
sale_state= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by ='Amount' , ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(x='State',y= 'Amount', data= sale_state )
```

NOTE-
from about we can say most of the order are placed from UP,Maharashtra and Karnatka

```python
#MARRITAL STATUS
sale_gen= df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by ='Amount' , ascending=False)
sns.set(rc={'figure.figsize':(7,6)})
sns.barplot(hue='Gender',x='Marital_Status',y= 'Amount', data= sale_gen )

```

MARRIED WOMEN

```python
sns.set(rc={'figure.figsize':(20,6)})
ax=sns.countplot(x='Occupation', data= df)
```

```python
sale_occ= df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by ='Amount' , ascending=False)
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(x='Occupation',y= 'Amount', data= sale_occ )
```

highest buyers are from IT AND healtecare

```python
#products
sns.set(rc={'figure.figsize':(27,6)})
ax=sns.countplot(x='Product_Category', data= df)
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
sale_pc= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by ='Amount' , ascending=False)
sns.set(rc={'figure.figsize':(30,6)})
sns.barplot(x='Product_Category',y= 'Amount', data= sale_pc )
```
